```python
#2
duplicates = df.duplicated().sum()
summary_stats = df.describe(include = 'all')

print(duplicates)
print(summary_stats)
```

```python
#3
survival_rate = df.groupby(['Pclass', 'Sex'])['Survived'].mean().reset_index()
survival_rate.columns = ['Pclass', 'Sex', 'Survival Rate']
print(survival_rate)
```

```python
#4
bins = [0, 12, 20, 40, 60, 80, 100]
labels = ['Child', 'Teen', 'Adult', 'Middle-Aged', 'Senior', 'Elderly']
df['Age Group'] = pd.cut(df['Age'], bins = bins, labels = labels, right = False) 
#DataFrame will have a new column called 'Age Group' that categorizes each passenger's age into one of the defined age groups.
age_survival_rate = df.groupby('Age Group', observed = False)['Survived'].mean()
print(age_survival_rate)
```

```python
#5
df['Family Size'] = df['SibSp'] + df['Parch'] + 1
family_survival_rate = df.groupby('Family Size')['Survived'].mean().reset_index()
correlation = df['Family Size'].corr(df['Survived'])

print(correlation)
print(family_survival_rate)
```

```python
#6
survivors_df = df[df['Survived'] == 1]
embarkation_counts = survivors_df['Embarked'].value_counts()
total_counts = df['Embarked'].value_counts()
survival_rates = (embarkation_counts / total_counts).reset_index()
survival_rates.columns = ['Embarked', 'Survival Rate']

most_common_embarkation = embarkation_counts.idxmax()
highest_survival_rate = survival_rates.loc[survival_rates['Survival Rate'].idxmax()]

print("Most common embarkation: ", most_common_embarkation)
print(survival_rates)
print(highest_survival_rate)
```

```python
#7
correlation = df['Fare'].corr(df['Survived'])
print("Correlation coefficient between Fare and Survival:", correlation)
```

The Supermarket Sales Dataset is a comprehensive dataset designed to provide insights into the sales performance of a supermarket.

We want to store this data in a relational database, with the following table schemas:

# Tables

## 1. Branches Table

| Column   | Type  | Key      | Description                              |
|----------|-------|----------|------------------------------------------|
| BranchID | INTEGER  | PRIMARY  | Unique identifier for each branch        |
| Branch | TEXT  |            |    Name of the supermarket branch        |
| City     | TEXT  |          | City where the branch is located         |



## 2. Products Table

| Column      | Type  | Key      | Description                               |
|-------------|-------|----------|-------------------------------------------|
| ProductID | INTEGER  | PRIMARY  | Unique identifier for each product line   |
| ProductLine | TEXT  |          | Category of the product |



## 3. Orders Table

| Column        | Type  | Key      | Description                                    |
|---------------|-------|----------|------------------------------------------------|
| OrderID       | TEXT  | PRIMARY  | Unique identifier for each order               |
| BranchID      | INTEGER  |   FK       | Identifier for the branch of the order    |
| ProductID   | INTEGER  |      FK    | Identifier linking to Products table ordered                      |
| Quantity      | INTEGER |        | Number of units ordered                        |
| UnitPrice   | REAL  |          | Price per unit of the product             |Total         | REAL  |          | Total sales value for the order               |
| OrderDate     | TEXT  |          | Date when the order was placed                |
| PaymentMethod | TEXT  |          | Method used for payment                       |
| CustomerType  | TEXT  |          | Type of customer (e.g., Member, Normal)     |
| Gender        | TEXT  |          | Gender of the customer                        |



Summary of Table Structures

* Branches Table contains information about different supermarket branches.

* Products Table details the various products available for purchase.
* Orders Table records the transactions made, linking products and branches to specific purchases.


# **Questions** **(35 Pts)**

1. Read the csv file and create this DB **(10 Pts)**
2. Write a query to find the top 3 product lines with the highest total quantity sold, using a LIMIT statement. **(4 Pts)**
3. Find the total revenue generated for each branch by product line **(4 Pts)**
4. Find the total sales for each branch **(3 Pts)**
5. Determine the number of orders placed by each gender **(3 Pts)**
6. List the product lines sold in each branch and the number of orders **(4 Pts)**
7. Find the highest quantity sold for a single order and the corresponding product line **(4 Pts)**
8. Check for duplicated invoices **(3 Pts)**


```python
#1
df = pd.read_csv("C:\\Users\\bsrlo\\OneDrive\\Masaüstü\\supermarket_sales.csv")

# Create a connection to SQLite database or create one
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Create Branches Table
cursor.execute('''CREATE TABLE IF NOT EXISTS Branches (
                    BranchID INTEGER PRIMARY KEY,
                    Branch TEXT,
                    City TEXT
                );''')

# Create Products Table
cursor.execute('''CREATE TABLE IF NOT EXISTS Products (
                    ProductID INTEGER PRIMARY KEY AUTOINCREMENT,
                    ProductLine TEXT
                );''')

# Create Orders Table
cursor.execute('''CREATE TABLE IF NOT EXISTS Orders (
                    OrderID TEXT PRIMARY KEY,
                    BranchID INTEGER,
                    ProductID INTEGER,
                    Quantity INTEGER,
                    UnitPrice REAL,
                    OrderDate TEXT,
                    PaymentMethod TEXT,
                    CustomerType TEXT,
                    Gender TEXT,
                    FOREIGN KEY (BranchID) REFERENCES Branches (BranchID),
                    FOREIGN KEY (ProductID) REFERENCES Products (ProductID)
                );''')

# Insert unique branches into Branches table
branches = df[['Branch', 'City']].drop_duplicates().reset_index(drop=True)

# This line assigns a unique ID to each branch based on its index. The + 1 ensures that IDs start from 1 instead of 0.
branches['BranchID'] = branches.index + 1 
branches.to_sql('Branches', conn, if_exists='replace', index=False)

# Insert unique product lines into Products table
products = df[['Product line']].drop_duplicates().reset_index(drop=True)
products['ProductID'] = products.index + 1  # Assign unique ProductIDs
products.columns = ['ProductLine', 'ProductID']
products.to_sql('Products', conn, if_exists='replace', index=False)

# Merge the data on 'Branch' and 'City' for BranchID and on 'Product line' for ProductID
merged_data = pd.merge(df, branches, on=['Branch', 'City'], how='left')
merged_data = pd.merge(merged_data, products, left_on='Product line', right_on='ProductLine', how='left')

# Prepare the Orders table data with proper column names
orders = merged_data[['Invoice ID', 'BranchID', 'ProductID', 'Quantity', 'Unit price', 'Date', 'Payment', 'Customer type', 'Gender']]
orders.columns = ['OrderID', 'BranchID', 'ProductID', 'Quantity', 'UnitPrice', 'OrderDate', 'PaymentMethod', 'CustomerType', 'Gender']
orders.to_sql('Orders', conn, if_exists='replace', index=False)

# Commit and close the connection
conn.commit()
conn.close()
```

```python
#2
# Connect to the SQLite database
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Find the top 3 product lines with the highest total quantity sold
query = """
SELECT P.ProductLine, SUM(O.Quantity) AS TotalQuantitySold
FROM Orders O
JOIN Products P ON O.ProductID = P.ProductID
GROUP BY P.ProductLine
ORDER BY TotalQuantitySold DESC
LIMIT 3;
"""

# Execute the query
cursor.execute(query)

# Fetch the results
results = cursor.fetchall()

# Display the results
for row in results:
    print(row)

# Close the connection
conn.close()

```

```python
#3
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Find the total revenue generated for each branch by product line
query = """
SELECT B.Branch, P.ProductLine, SUM(O.Quantity * O.UnitPrice) AS TotalRevenue
FROM Orders O
JOIN Branches B ON O.BranchID = B.BranchID
JOIN Products P ON O.ProductID = P.ProductID
GROUP BY B.Branch, P.ProductLine
ORDER BY B.Branch, TotalRevenue DESC;
"""

cursor.execute(query)

results = cursor.fetchall()

for row in results:
    print(row)

conn.close()
```

```python
#4
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Find the total sales for each branch
query = """
SELECT B.Branch, SUM(O.Quantity * O.UnitPrice) AS TotalSales
FROM Orders O
JOIN Branches B ON O.BranchID = B.BranchID
GROUP BY B.Branch
ORDER BY TotalSales DESC;
"""

cursor.execute(query)

total_sales_results = cursor.fetchall()

print("Total Sales for Each Branch:")
for row in total_sales_results:
    print(row)

conn.close()
```

```python
#5
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Determine the number of orders placed by each gender
query = """
SELECT Gender, COUNT(OrderID) AS NumberOfOrders
FROM Orders
GROUP BY Gender;
"""

cursor.execute(query)

gender_orders_results = cursor.fetchall()

print("\nNumber of Orders Placed by Each Gender:")
for row in gender_orders_results:
    print(row)

conn.close()
```

```python
#6
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# List the product lines sold in each branch and the number of orders
query = """
SELECT B.Branch, P.ProductLine, COUNT(O.OrderID) AS NumberOfOrders
FROM Orders O
JOIN Branches B ON O.BranchID = B.BranchID
JOIN Products P ON O.ProductID = P.ProductID
GROUP BY B.Branch, P.ProductLine
ORDER BY B.Branch, NumberOfOrders DESC;
"""

cursor.execute(query)

product_lines_results = cursor.fetchall()

print("\nProduct Lines Sold in Each Branch and Number of Orders:")
for row in product_lines_results:
    print(row)

conn.close()
```

```python
#7
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Find the highest quantity sold for a single order and the corresponding product line
query = """
SELECT O.OrderID, P.ProductLine, MAX(O.Quantity) AS MaxQuantity
FROM Orders O
JOIN Products P ON O.ProductID = P.ProductID
GROUP BY O.OrderID
ORDER BY MaxQuantity DESC
LIMIT 1;
"""

cursor.execute(query)

highest_quantity_results = cursor.fetchall()

print("\nHighest Quantity Sold for a Single Order:")
for row in highest_quantity_results:
    print(row)

conn.close()
```

```python
#8
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()

# Check for duplicated invoices
query = """
SELECT OrderID, COUNT(OrderID) AS DuplicateCount
FROM Orders
GROUP BY OrderID
HAVING DuplicateCount > 1;
"""

cursor.execute(query)

duplicates_results = cursor.fetchall()

print("\nDuplicated Invoices:")
for row in duplicates_results:
    print(row)

conn.close()
```
